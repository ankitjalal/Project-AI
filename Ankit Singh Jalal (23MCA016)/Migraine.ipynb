## Creating a Pipeline

```python
#Pipeline we made is Organized the data according to the steps that given to them 
#Here we give Simple imputer And standard scaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    #     ..... add as many as you want in your pipeline
    ('std_scaler', StandardScaler()),
])
```

```python
#here we use migraine original dataset 
#because we apply imputer in our pipeline no need to add migraine_tr (here we apply imputer above)
migraine_num_tr = my_pipeline.fit_transform(migraine)
```

```python
migraine_num_tr.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(migraine_num_tr, migraine_labels)
```

```python
some_data = migraine.iloc[:5]
```

```python
some_labels = migraine_labels.iloc[:5]
```

```python
prepared_data = my_pipeline.transform(some_data)
```
